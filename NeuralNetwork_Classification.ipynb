```python
from sklearn.datasets import make_circles
n_samples = 1000
X, y = make_circles(n_samples,
                    noise=0.03,
                    random_state = 42)

print(f"First X features:\n{X[:5]}")
print(f"First Y features:\n{y[:5]}")
```

```python
#make data frame of Circle data
import pandas as pd
circles = pd.DataFrame({"X1": X[:,0],"X2": X[:,1], "label": y});
circles.head(100)
```
